    **Notes:**

    * Plots 2 - 5 isolate the above observations for clarity. 
    

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cdf = pd.read_csv(city_data_to_load)  # City data 
rdf = pd.read_csv(ride_data_to_load)  # Ride data

# Combine the data into a single dataset
rdat = pd.merge(rdf, cdf, on='city', how='left')

# Display the data table for preview
rdat.head()
```

## Bubble Plot of Ride Sharing Data   (Plot 1)
* Plot the Bubble Chart and save plot image to a file.
* Observe the following :
    * Average Fare amount is higher in non-Urban cities, relative to Urban cities.
    * Number of rides is significantly higher in Urban cities, relative to non-Urban cities.
    * Number of drivers in Urban cities, seems to be higher than in non-Urban cities.
    

```python
# Prepare data grouped by City Name with aggregates for Fare Average and Ride Count
sumdf = rdat.groupby('city').agg({'fare':'mean','ride_id':'count', 'driver_count':'first', 'type':'first'})

# Common parameters for each scatter plot data set
sctargs = {'alpha':0.7,'edgecolors':'black', 'label':None}
colors = {'Rural':'gold','Urban':'lightcoral','Suburban':'lightskyblue'}

# Set size for the plot figure
plt.figure(figsize=(8,5))

# Plot data for each city type
for type in ['Rural', 'Urban', 'Suburban'] :
    data = sumdf[sumdf['type']==type]                # Ride data for city type
    plt.scatter([], [], c=colors[type], label=type)  # Dummy plot to fix size of legend marker
    plt.scatter(data['ride_id'], data['fare'], s=10*data['driver_count'], c=colors[type], **sctargs)  # Plot data
    
# Place legend, axis labels
plt.legend(loc='upper right')
plt.ylabel('Average Fare ($)')
plt.xlabel('Total Number of Rides per City')

# Define axis limits
plt.xlim(0, sumdf['ride_id'].max()+3, 5)
plt.ylim(sumdf['fare'].min()-3, sumdf['fare'].max()+3, 5)

# Add title and text box with note regarding bubble size
plt.title('Pyber Ride Sharing Data (2016)'); plt.grid(b=True)
plt.text(43,35,'Note: \nCircle size \nrepresents \nnumber of \ndrivers \navailable \nper city',
         bbox={'facecolor':'lightgrey', 'alpha':0.2})

plt.savefig('Images/PyberRideShare2016_BubblePlot.png')
```

## Prepare data for Percentage vs City Type Plots
* Aggregate the ride/city data by City Type for : 
    * Total Fare amount
    * Total Ride count
    * Total Driver count

```python
# Group data by City Type and aggregate Total Fare, Total Ride Count and Total Driver Count
typedf = rdat.groupby('type').agg({'fare':'sum','ride_id':'count'})
typedf = typedf.join(cdf.groupby('type')['driver_count'].sum())

# Map colors and pie slice to City type
piecolor = [colors[type] for type in typedf.index]
piecut = [{'Rural':0, 'Urban':0.1, 'Suburban':0}[type] for type in typedf.index]

# Capture common plot parameters for Percentage vs City Type Pie Charts
pctargs = {'labels':typedf.index, 'colors':piecolor, 'startangle':150, 
         'shadow':True, 'autopct':'%1.1f%%', 'explode':piecut}
```

## Percentage Total Fares by City Type  (Plot 2)

* Plot Pie Chart for percentages and save plot image to a file.
* Observe that Total Fares is significantly higher in Urban cities, relative to non-Urban cities.

```python
# Set size for the plot figure
plt.figure(figsize=(8,7))

# Pie Chart for Percentage Total Fares by City Type
plt.pie(typedf['fare'], **pctargs)
plt.title('% of Total Fares by City Type')

#Save plot figure to file : PctTotFarebyCityType.png
plt.savefig('Images/PctTotFarebyCityType_PieChart.png')
```

## Percentage Total Rides by City Type  (Plot 3)
* Plot Pie Chart for percentages and save plot image to a file.
* Observe that Total number of rides is significantly higher in Urban areas, relative to non-Urban areas.

```python
# Set size for the plot figure
plt.figure(figsize=(8,7))

# Pie Chart for Percentage Total Fares by City Type
plt.pie(typedf['ride_id'], **pctargs)
plt.title('% of Total Rides by City Type')

#Save plot figure to file : PctTotFarebyCityType.png
plt.savefig('Images/PctTotRidesbyCityType_PieChart.png')
```

## Total Drivers by City Type  (Plot 4)
* Plot Pie Chart for percentages and save plot image to a file.
* Observe that number of drivers in Urban cities is significantly higher that non-urban cities.
    * The number of available drivers decreases as distance from Urban areas increases.
    * This may indicate fewer ride share opportunites in non-Urban areas.
    * This may indicate fewer drivers available as a prcentage of the population.

```python
# Set size for the plot figure
plt.figure(figsize=(8,7))

# Pie Chart for Percentage Total Fares by City Type
plt.pie(typedf['driver_count'], **pctargs)
plt.title('% of Total Drivers by City Type')

#Save plot figure to file : PctTotFarebyCityType.png
plt.savefig('Images/PctTotDriversbyCityType_PieChart.png')
```

## Fare Values by City Type   (Plot 5)
* Plot Fare Values by City Type and save plot image to a file.
* Observe that Fare Values, on averge, are higher for non-Urban cities, particulary for Rural cites. 
    * This may indicate that distance per ride is related to distance from an Urban area.
    * This may indicate higher fare values due to demand relative to fewer drivers in non-Urban areas

```python
# By City Type, get Average, Min and Max for Fare Values
fareavg = rdat.groupby('type')['fare'].mean()
faremin = rdat.groupby('type')['fare'].min()
faremax = rdat.groupby('type')['fare'].max()

# Define aesthetics for the bars. 
# Error boars are used to show range of Fare Amounts i.e. min/max
barcolor = [colors[type] for type in fareavg.index]
pltargs  = {'color':barcolor, 'width':.4, 'align':'center', 'edgecolor':'grey', 'linewidth':3}
rngargs  = {'ecolor':'olive', 'capsize':5}

# Bar chart for Fare Values by City Type
plt.figure(figsize=(8,6))
plt.bar(fareavg.index, fareavg, yerr=[fareavg-faremin, faremax-fareavg], **pltargs, error_kw=rngargs)
plt.title('Average Fare by City Type')
plt.xlabel('City Type')
plt.ylabel('Average Fare Amount ($)')
plt.text(2.4,40,'Note: \nError bars \nindicate range \nof Fare value \ni.e Min/Max',
         bbox={'facecolor':'lightgrey', 'alpha':0.2})

#Save plot figure to file : FareValuebyCityType.png
plt.savefig('Images/FareValuebyCityType_BarChart.png')
```
